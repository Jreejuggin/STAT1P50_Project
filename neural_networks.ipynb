```python
from tensorflow.keras.layers import Dense, InputLayer
import pandas as pd
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
import tensorflow as tf
tf.random.set_seed(1)

yList = [] #empty list
#the model
model = Sequential()

# input and (4) hidden layers
input_layer = InputLayer(input_shape=(5,))
model.add(input_layer)
hidden_layer = Dense(3)
model.add(hidden_layer)

hiddenLayer2 = Dense(4)
model.add(hiddenLayer2)

hiddenLayer3 = Dense (5)
model.add(hiddenLayer3)

hiddenLayer4 = Dense (7)
model.add(hiddenLayer4)

output_layer = Dense(1, activation='sigmoid')
model.add(output_layer)

#after creating model configure it
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'])



 
data = pd.read_csv('Life Expectancy Data.csv')
# features
X = data[['Adult Mortality',' BMI ',' HIV/AIDS','Income composition of resources','Schooling']]

# target variable
y = data['Life expectancy ']
#inserts(into list)life expectancy data into categories (0 to 1)
for i in range(len(y)):
    if y[i] >= 70: #the mean life expectancy
        y[i] = 1 #if life expectancy is above 70, it is 1(high)
        yList.append(1)
    else:
        y[i] = 0 #if its below 70 it is 0(low)
        yList.append(0)
        
#training of the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

model.fit(X_train, y_train, epochs=10)

results = model.evaluate(X_test, y_test)
print("test loss, test acc:", results)


print("Number of weights after calling the model:", len(model.weights))
```
