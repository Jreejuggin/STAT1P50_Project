```python
import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator, BayesianEstimator
from pgmpy.inference import VariableElimination
from sklearn.base import BaseEstimator

# Load data from CSV file
data = pd.read_csv("Life Expectancy Data.csv")

#turning the data types to int
data['Life expectancy '] = data['Life expectancy '].astype('int')
data[' HIV/AIDS'] = data[' HIV/AIDS'].astype('int')
data[' BMI '] = data[' BMI '].astype('int') 
data['Income composition of resources'] = data['Income composition of resources'].astype('int')
data[ 'Schooling'] = data[ 'Schooling'].astype('int')


data.to_csv("Life Expectancy Data.csv", index=False)

#creating the networks 
model = BayesianNetwork([('Life expectancy ','Adult Mortality'),('Life expectancy ',' BMI '),
                         ('Life expectancy ',' HIV/AIDS'),
                         ('Life expectancy ','Income composition of resources'),
                         ('Life expectancy ','Schooling')])

#get the maximum likelihood to get the condiitonal probability 
model.fit(data, estimator= MaximumLikelihoodEstimator)

#print out the probability of each feature affecting the life expectancy 
print(model.get_cpds('Adult Mortality')) 
print(model.get_cpds(' BMI ')) 
print(model.get_cpds(' HIV/AIDS')) 
print(model.get_cpds('Income composition of resources')) 
print(model.get_cpds('Schooling'))
```
